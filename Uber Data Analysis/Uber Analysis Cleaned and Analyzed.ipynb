```python
import pandas as pd
```

```python
df = pd.read_csv("../Dataset/UberDataset.csv")
df.head(10)
```

```python
df.tail(10)
```

```python
#Data Cleaning
df = df[df['START_DATE'] != 'Totals'].copy()
df.tail()
```

```python
df.columns
```

```python
df.info()
```

```python
df['PURPOSE'] = df['PURPOSE'].fillna('Unknown')
df['START_DATE'] = df['START_DATE'].str.strip()
df['END_DATE']   = df['END_DATE'].str.strip()
df['CATEGORY'] = df['CATEGORY'].str.strip()
df['START']   = df['START'].str.strip()
df['STOP']   = df['STOP'].str.strip()
df['PURPOSE']   = df['PURPOSE'].str.strip()
df.head()
```

```python
print("Missing values per column:")
print(df.isnull().sum())
```

```python
df['START_DATE'] = df['START_DATE'].str.replace('/', '-')
df['END_DATE']   = df['END_DATE'].str.replace('/', '-')
df['START_DATE'] = pd.to_datetime(df['START_DATE'], dayfirst=False)
df['END_DATE']   = pd.to_datetime(df['END_DATE'], dayfirst=False)
```

```python
df.dtypes
```

```python
#Data Analysis
#Total Trips
total = len(df)
print("Total Trips : ",total)
# Trips by Category
purpose_counts = df['PURPOSE'].value_counts()
Percentage_purpose_counts = (purpose_counts/len(df) * 100).round(2).sort_values(ascending=False)
print("\nPercentage of Trips by Purpose : ")
print(Percentage_purpose_counts)
# Trips by Purpose
category_counts = df['CATEGORY'].value_counts()
Percentage_category_counts = (category_counts/len(df) * 100).round(2).sort_values(ascending=False)
print("\nPercentage of Trips by Category : ")
Percentage_category_counts
```

```python
# Total distance covered by all trips
total_distance = df['MILES'].sum()
print("Total Distance Covered:", total_distance)
# Average distance per trip
avg_distance = df['MILES'].mean().round(2)
print("Average Distance per Trip:", avg_distance)
# Shortest trip
min_distance = df['MILES'].min()
print("Shortest Trip:", min_distance)
# Longest trip
max_distance = df['MILES'].max()
print("Longest Trip:", max_distance)
#Average distance per purpose
avg_distance_by_purpose = df.groupby('PURPOSE')['MILES'].mean().round(2).sort_values(ascending=False)
print("\nAverage Distance per Purpose : ")
print(avg_distance_by_purpose)
```

```python
#Peak Hours for Rides
peak_hour = df['START_DATE'].dt.hour.value_counts().sort_values(ascending=False)
print("Trips per Hour: ", peak_hour)
```

```python
#Frequency of Pickup Locations
pickup_locations = df['START'].value_counts().sort_values(ascending=False)
print(pickup_locations)
```

```python
#Frequency of Routes
df['ROUTE'] = df['START'] + " → " + df['STOP']
df['ROUTE']   = df['ROUTE'].str.strip()
route_counts = df['ROUTE'].value_counts().sort_values(ascending=False)
print(route_counts)
```

```python
df.to_csv("Uber Dataset Cleaned and Analyzed.csv", index=False)
```
